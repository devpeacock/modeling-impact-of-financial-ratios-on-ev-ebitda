# ANOVA and Cluster Analysis: Economic Interpretation of Results

This notebook focuses on what the results tell us about company types and their EV/EBITDA valuations.

Interpretation goal: translate cluster profiles into clear business archetypes (growth, profitability, leverage, capital intensity, innovation) and show how these archetypes drive differences in valuation multiples.

## Data Loading and Preparation

This notebook uses only three prepared datasets:
- `df_sectors_cook2.csv` (source for clustering)
- `ml_train_dataset_used_for_models.csv`
- `ml_test_dataset_used_for_models.csv`

A lightweight preparation step maps columns to the variables used in ANOVA sections.

```python
def filter_groups_with_min_size(
    data: pd.DataFrame,
    group_column: str,
    min_size: int = MIN_GROUP_SIZE,
) -> pd.DataFrame:
    return (
        data.dropna(subset=[group_column])
        .groupby(group_column, group_keys=False)
        .filter(lambda frame: len(frame) >= min_size)
    )


def load_input_datasets() -> tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame, pd.DataFrame]:
    """Load fixed processed datasets used in this project."""
    df_sectors_cook2 = pd.read_csv(Path("../data/processed/df_sectors_cook2.csv"))
    ml_train = pd.read_csv(Path("../data/processed/ml_train_dataset_used_for_models.csv"))
    ml_test = pd.read_csv(Path("../data/processed/ml_test_dataset_used_for_models.csv"))

    sector_cols = [col for col in df_sectors_cook2.columns if col.startswith("Sector_")]
    base_sector = "Sector_Base"

    sector_labels = pd.Series(base_sector, index=df_sectors_cook2.index, name="sector")
    if sector_cols:
        dummy_frame = df_sectors_cook2[sector_cols].astype(bool)
        has_dummy = dummy_frame.any(axis=1)
        sector_labels.loc[has_dummy] = (
            df_sectors_cook2.loc[has_dummy, sector_cols]
            .idxmax(axis=1)
            .str.replace("Sector_", "", regex=False)
        )

    df_sectors_cook2 = df_sectors_cook2.copy()
    df_sectors_cook2["row_id"] = df_sectors_cook2.index.astype(int)
    df_sectors_cook2["sector"] = sector_labels
    df_sectors_cook2["company"] = "obs_" + df_sectors_cook2["row_id"].astype(str)

    # Lightweight mapping for ANOVA sections kept in this notebook.
    prepared = pd.DataFrame(
        {
            "row_id": df_sectors_cook2["row_id"],
            "company": df_sectors_cook2["company"],
            "sector": df_sectors_cook2["sector"],
            "revenue_growth": df_sectors_cook2["Revenue_Growth"],
            "return_on_equity": df_sectors_cook2["returnOnEquity"],
            "ebitda_margin": df_sectors_cook2["Profit_Margin"],
            "debt_to_equity_ratio": df_sectors_cook2["Debt_to_Equity"],
            "rd_to_revenue": df_sectors_cook2["R_and_D_to_Revenue"],
            "capex_to_revenue": df_sectors_cook2["Capex_to_Revenue"],
            "price_earnings_ratio": df_sectors_cook2["Enterprise_Value_over_EBITDA"],
        }
    )

    prepared["revenue_growth_percent"] = prepared["revenue_growth"] * 100
    prepared["return_on_equity_percent"] = prepared["return_on_equity"] * 100
    prepared["ebitda_margin_percent"] = prepared["ebitda_margin"] * 100
    prepared["rd_to_revenue_percent"] = prepared["rd_to_revenue"] * 100
    prepared["capex_to_revenue"] = prepared["capex_to_revenue"] * 100

    return df_sectors_cook2, ml_train, ml_test, prepared
```

## Visualization Helpers

Reusable functions for histograms, Q-Q plots, and sector boxplots.

```python
def plot_histogram(
    values: pd.Series,
    title: str,
    x_label: str,
    x_limits: tuple[float, float] | None,
    bins: int,
) -> None:
    plt.figure(figsize=(8, 4))
    plt.hist(values.dropna(), bins=bins, color="lightblue", edgecolor="black")
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel("Frequency")
    if x_limits is not None:
        plt.xlim(*x_limits)
    plt.tight_layout()
    plt.show()


def plot_qq(values: pd.Series, title: str) -> None:
    plt.figure(figsize=(6, 6))
    stats.probplot(values.dropna(), dist="norm", plot=plt)
    plt.title(title)
    plt.tight_layout()
    plt.show()


def plot_sector_boxplot(
    data: pd.DataFrame,
    value_column: str,
    title: str,
    y_label: str,
    y_limits: tuple[float, float] | None,
) -> None:
    plt.figure(figsize=(12, 5))
    sns.boxplot(
        data=data,
        x="sector",
        y=value_column,
        palette="Blues",
        showfliers=False,
    )
    plt.title(title)
    plt.xlabel("Sectors")
    plt.ylabel(y_label)
    plt.xticks(rotation=90)
    if y_limits is not None:
        plt.ylim(*y_limits)
    plt.tight_layout()
    plt.show()
```

## Statistical Test Helpers

- `run_kruskal_wallis_test` - non-parametric one-way test of equality of group medians.
- `compute_shapiro_by_sector` - per-sector Shapiro-Wilk normality test (skips sectors with n < 3).

```python
def run_kruskal_wallis_test(data: pd.DataFrame, value_column: str, group_column: str) -> tuple:
    groups = [
        group[value_column].dropna().to_numpy()
        for _, group in data.groupby(group_column)
    ]
    groups = [group for group in groups if len(group) > 0]

    if len(groups) < 2:
        raise ValueError("Kruskal-Wallis test requires at least two non-empty groups.")

    return stats.kruskal(*groups)


def compute_shapiro_by_sector(data: pd.DataFrame, value_column: str) -> pd.DataFrame:
    results: list[dict[str, object]] = []

    for sector, group in data.groupby("sector"):
        values = group[value_column].dropna()
        if len(values) >= 3:
            with warnings.catch_warnings():
                warnings.simplefilter("ignore")
                p_value = stats.shapiro(values).pvalue
            results.append({"sector": sector, "p_value": p_value})

    return pd.DataFrame(results)
```

## Clustering Helpers

Methods for determining the optimal number of K-Means clusters:
- **Elbow method** (within-cluster sum of squares)
- **Silhouette method** (average silhouette score)
- **Gap statistic** (comparison to uniform reference distribution)

```python
def plot_elbow_method(scaled_data: np.ndarray, max_clusters: int = MAX_CLUSTERS) -> None:
    wss_values: list[float] = []

    for clusters in range(1, max_clusters + 1):
        model = KMeans(n_clusters=clusters, n_init=25, random_state=RANDOM_SEED)
        model.fit(scaled_data)
        wss_values.append(model.inertia_)

    plt.figure(figsize=(8, 4))
    plt.plot(range(1, max_clusters + 1), wss_values, marker="o")
    plt.title("Elbow Method - Optimal Number of Clusters")
    plt.xlabel("Number of clusters")
    plt.ylabel("Within-cluster sum of squares")
    plt.tight_layout()
    plt.show()


def plot_silhouette_method(scaled_data: np.ndarray, max_clusters: int = MAX_CLUSTERS) -> None:
    silhouette_values: list[float] = []
    cluster_range = range(2, max_clusters + 1)

    for clusters in cluster_range:
        model = KMeans(n_clusters=clusters, n_init=25, random_state=RANDOM_SEED)
        labels = model.fit_predict(scaled_data)
        silhouette_values.append(silhouette_score(scaled_data, labels))

    plt.figure(figsize=(8, 4))
    plt.plot(list(cluster_range), silhouette_values, marker="o")
    plt.title("Silhouette Method")
    plt.xlabel("Number of clusters")
    plt.ylabel("Average silhouette score")
    plt.tight_layout()
    plt.show()


def compute_gap_statistic(
    scaled_data: np.ndarray,
    max_clusters: int = MAX_CLUSTERS,
    reference_runs: int = 50,
) -> pd.DataFrame:
    rng = np.random.default_rng(RANDOM_SEED)
    mins = scaled_data.min(axis=0)
    maxs = scaled_data.max(axis=0)

    rows: list[dict[str, float]] = []

    for clusters in range(1, max_clusters + 1):
        model = KMeans(n_clusters=clusters, n_init=25, random_state=RANDOM_SEED)
        model.fit(scaled_data)
        observed_dispersion = model.inertia_

        reference_dispersions: list[float] = []
        for _ in range(reference_runs):
            reference_data = rng.uniform(low=mins, high=maxs, size=scaled_data.shape)
            reference_model = KMeans(n_clusters=clusters, n_init=25, random_state=RANDOM_SEED)
            reference_model.fit(reference_data)
            reference_dispersions.append(reference_model.inertia_)

        log_reference = np.log(reference_dispersions)
        gap_value = log_reference.mean() - np.log(observed_dispersion)
        sk = np.sqrt(1 + 1 / reference_runs) * log_reference.std(ddof=1)

        rows.append({"k": float(clusters), "gap": float(gap_value), "sk": float(sk)})

    return pd.DataFrame(rows)


def plot_gap_statistic(gap_df: pd.DataFrame) -> None:
    plt.figure(figsize=(8, 4))
    plt.errorbar(gap_df["k"], gap_df["gap"], yerr=gap_df["sk"], fmt="-o")
    plt.title("Gap Statistic")
    plt.xlabel("Number of clusters")
    plt.ylabel("Gap value")
    plt.tight_layout()
    plt.show()
```

## Cluster Statistics Helpers

- `generate_cluster_statistics_flat` - one wide row per cluster with mean/median/min/max/variance for each variable (for quick printing).
- `generate_cluster_statistics_table` - long-format table suitable for CSV export.

```python
def generate_cluster_statistics_flat(
    clean_data: pd.DataFrame,
    cluster_number: int,
    analysis_variables: list[str],
) -> pd.DataFrame:
    subset = clean_data[clean_data["cluster"] == cluster_number]

    row: dict[str, float] = {}
    for variable in analysis_variables:
        values = subset[variable]
        row[f"{variable}_mean"] = values.mean(skipna=True)
        row[f"{variable}_median"] = values.median(skipna=True)
        row[f"{variable}_min"] = values.min(skipna=True)
        row[f"{variable}_max"] = values.max(skipna=True)
        row[f"{variable}_variance"] = values.var(skipna=True)

    return pd.DataFrame([row])


def generate_cluster_statistics_table(
    clean_data: pd.DataFrame,
    cluster_number: int,
    analysis_variables: list[str],
) -> pd.DataFrame:
    subset = clean_data[clean_data["cluster"] == cluster_number]
    summary = subset[analysis_variables].agg(["mean", "median", "min", "max", "var"]).T.reset_index()
    summary.columns = ["Variable", "Mean", "Median", "Min", "Max", "Variance"]
    return summary
```

## Cluster-Filtered ML Pipeline and Comparison with Tuned XGBoost (from notebook 03)
After clustering, this section maps cluster labels to ml_train/ml_test using row_id, removes clusters that are too small for tuned XGBoost, trains a per-cluster model, and compares it against the global tuned XGBoost baseline.

```python
def regression_metrics(model_name: str, y_true: pd.Series, y_pred: np.ndarray) -> dict[str, float | str]:
    y_true_arr = np.asarray(y_true, dtype=float)
    y_pred_arr = np.asarray(y_pred, dtype=float)
    eps = 1e-9
    mape_value = float(np.mean(np.abs((y_true_arr - y_pred_arr) / np.maximum(np.abs(y_true_arr), eps))) * 100)
    return {
        "Model": model_name,
        "R2": float(r2_score(y_true_arr, y_pred_arr)),
        "MAE": float(mean_absolute_error(y_true_arr, y_pred_arr)),
        "RMSE": float(np.sqrt(mean_squared_error(y_true_arr, y_pred_arr))),
        "MAPE (%)": mape_value,
    }


best_xgb_params_03 = {
    "n_estimators": 387,
    "max_depth": 4,
    "learning_rate": 0.02408322616990878,
    "subsample": 0.7434587125184655,
    "colsample_bytree": 0.9376852562905246,
    "gamma": 0.9300168348108319,
    "reg_lambda": 3.355717584120253,
}

target_col = "Enterprise_Value_over_EBITDA"
drop_cols = {target_col, "split", "Sector_label", "row_id", "cluster", "company"}

cluster_lookup = clean_data[["row_id", "company", "cluster"]].copy()
cluster_lookup["row_id"] = pd.to_numeric(cluster_lookup["row_id"], errors="coerce").astype("Int64")

ml_train_clustered = ml_train_df.copy()
ml_test_clustered = ml_test_df.copy()
ml_train_clustered["row_id"] = pd.to_numeric(ml_train_clustered["row_id"], errors="coerce").astype("Int64")
ml_test_clustered["row_id"] = pd.to_numeric(ml_test_clustered["row_id"], errors="coerce").astype("Int64")

ml_train_clustered = ml_train_clustered.merge(cluster_lookup, on="row_id", how="left")
ml_test_clustered = ml_test_clustered.merge(cluster_lookup, on="row_id", how="left")

cluster_sizes_train = ml_train_clustered["cluster"].value_counts(dropna=True).sort_index()
valid_clusters = cluster_sizes_train[cluster_sizes_train >= MIN_CLUSTER_TRAIN_SIZE_FOR_XGB].index.tolist()

if not valid_clusters:
    raise ValueError(
        "No cluster reaches MIN_CLUSTER_TRAIN_SIZE_FOR_XGB. Lower threshold or use a fallback global model only."
    )

filtered_train = ml_train_clustered[ml_train_clustered["cluster"].isin(valid_clusters)].copy()
filtered_test = ml_test_clustered[ml_test_clustered["cluster"].isin(valid_clusters)].copy()

# Filtered datasets kept for downstream ML blocks in this notebook.
ml_train_df_filtered = filtered_train.copy()
ml_test_df_filtered = filtered_test.copy()

removed_train = ml_train_clustered[~ml_train_clustered["cluster"].isin(valid_clusters)].copy()
removed_test = ml_test_clustered[~ml_test_clustered["cluster"].isin(valid_clusters)].copy()

feature_cols = [
    col for col in filtered_train.columns
    if col in filtered_test.columns and col not in drop_cols
]

if not feature_cols:
    raise ValueError("No features left after filtering columns for ML.")

X_train_filtered = filtered_train[feature_cols].copy()
y_train_filtered = filtered_train[target_col].copy()
X_test_filtered = filtered_test[feature_cols].copy()
y_test_filtered = filtered_test[target_col].copy()

if X_train_filtered.empty or X_test_filtered.empty:
    raise ValueError("Filtered train/test is empty after removing small clusters.")

global_tuned_xgb = xgb.XGBRegressor(
    objective="reg:squarederror",
    random_state=RANDOM_SEED,
    n_jobs=-1,
    **best_xgb_params_03,
    )
global_tuned_xgb.fit(X_train_filtered, y_train_filtered)
y_pred_global_tuned = global_tuned_xgb.predict(X_test_filtered)

cluster_models: dict[int, object] = {}
cluster_model_rows: list[dict[str, object]] = []
for cluster_id in sorted(valid_clusters):
    cluster_train = filtered_train[filtered_train["cluster"] == cluster_id]
    X_cluster_train = cluster_train[feature_cols]
    y_cluster_train = cluster_train[target_col]

    if len(cluster_train) >= MIN_CLUSTER_TRAIN_SIZE_FOR_XGB:
        model_local = xgb.XGBRegressor(
            objective="reg:squarederror",
            random_state=RANDOM_SEED,
            n_jobs=-1,
            **best_xgb_params_03,
            )
        model_label = "XGBoost (tuned from 03)"
    else:
        model_local = Ridge(alpha=1.0)
        model_label = "Ridge fallback"

    model_local.fit(X_cluster_train, y_cluster_train)
    cluster_models[int(cluster_id)] = model_local
    cluster_model_rows.append({
        "cluster": int(cluster_id),
        "train_size": int(len(cluster_train)),
        "model": model_label,
    })

y_pred_clustered = pd.Series(index=filtered_test.index, dtype=float)
for cluster_id in sorted(valid_clusters):
    test_mask = filtered_test["cluster"] == cluster_id
    if not bool(test_mask.any()):
        continue
    X_cluster_test = filtered_test.loc[test_mask, feature_cols]
    y_pred_clustered.loc[test_mask] = cluster_models[int(cluster_id)].predict(X_cluster_test)

clustered_metrics = regression_metrics(
    "Cluster-aware (separate model per cluster)",
    y_test_filtered,
    y_pred_clustered.loc[y_test_filtered.index].to_numpy(),
)
global_metrics = regression_metrics(
    "Global tuned XGBoost from 03",
    y_test_filtered,
    y_pred_global_tuned,
)

comparison_df = pd.DataFrame([clustered_metrics, global_metrics]).sort_values(by="R2", ascending=False).reset_index(drop=True)
cluster_model_df = pd.DataFrame(cluster_model_rows)

print("Train cluster sizes before filtering:")
print(cluster_sizes_train)
print("\nValid clusters kept for ML:")
print(valid_clusters)
print("\nRows removed from train/test due to too-small clusters:")
print({
    "train_removed": int(removed_train.shape[0]),
    "test_removed": int(removed_test.shape[0]),
})

if not removed_train.empty or not removed_test.empty:
    removed_companies = pd.concat([removed_train[["company"]], removed_test[["company"]]], axis=0)
    removed_companies = removed_companies.dropna().drop_duplicates().sort_values(by="company")
    print("\nRemoved companies (based on clustered row_id -> company mapping):")
    print(removed_companies.head(30))

print("\nPer-cluster model assignment:")
print(cluster_model_df)

print("\nModel comparison on filtered test set:")
print(comparison_df)

plt.figure(figsize=(9, 3.8))
sns.heatmap(comparison_df.set_index("Model")[["R2", "MAE", "RMSE", "MAPE (%)"]], annot=True, fmt=".3f", cmap="Blues")
plt.title("Cluster-aware vs tuned global XGBoost (from notebook 03)")
plt.tight_layout()
plt.show()
```

## What Does the Model Comparison Mean for Cluster Interpretation?

The out-of-sample test result shows that the global XGBoost model (R2 = 0.325) clearly outperforms the "separate model per cluster" approach (R2 = 0.236).

Economic interpretation:
- Clusters describe business regimes and valuation levels well, but they are not necessarily the best split for building separate predictive models.
- In practice, part of company-level information is shared across clusters, and the global model learns these cross-sectional relationships better.
- In this project, the strongest value of clustering is interpretive segmentation (risk/growth/valuation profile), not improvement of pure forecast accuracy.

## In-Depth Economic Interpretation of Clusters (Result -> Business Meaning)

This section emphasizes outcomes rather than technique.

We read clusters through three layers:
1. Financial profile: growth, profitability, leverage, R&D, and CAPEX versus the full-sample average.
2. Sector structure: whether a cluster is diversified or dominated by one business type.
3. Valuation: whether a cluster is rewarded with a premium or priced at a discount in EV/EBITDA.

Final practical takeaway: each cluster should be treated as a distinct company type with a different value-creation logic.

```python
from scipy.spatial.distance import pdist, squareform
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score, silhouette_samples

if "cluster" not in clean_data.columns:
    raise ValueError("Run the clustering fit cell first.")

analysis_df = clean_data.copy()
analysis_df["cluster"] = analysis_df["cluster"].astype(int)
cluster_labels = analysis_df["cluster"].to_numpy()

overall_mean = analysis_df[clustering_columns].mean()
overall_std = analysis_df[clustering_columns].std(ddof=0).replace(0, np.nan)

cluster_profile_mean = analysis_df.groupby("cluster")[clustering_columns].mean()
cluster_profile_median = analysis_df.groupby("cluster")[clustering_columns].median()
cluster_profile_z = (cluster_profile_mean - overall_mean) / overall_std
cluster_sizes = analysis_df["cluster"].value_counts().sort_index()

sil_values = silhouette_samples(scaled_cluster_data, cluster_labels)
analysis_df["silhouette_value"] = sil_values
silhouette_cluster_stats = analysis_df.groupby("cluster")["silhouette_value"].agg(["mean", "median", "min", "max"])
silhouette_cluster_stats["pct_negative"] = analysis_df.groupby("cluster")["silhouette_value"].apply(lambda s: float((s < 0).mean() * 100.0))

overall_silhouette = float(silhouette_score(scaled_cluster_data, cluster_labels))
calinski_harabasz = float(calinski_harabasz_score(scaled_cluster_data, cluster_labels))
davies_bouldin = float(davies_bouldin_score(scaled_cluster_data, cluster_labels))

centroids_scaled = pd.DataFrame(
    kmeans_model.cluster_centers_,
    index=sorted(np.unique(cluster_labels)),
    columns=clustering_columns,
)
centroid_distance = pd.DataFrame(
    squareform(pdist(centroids_scaled.to_numpy(), metric="euclidean")),
    index=centroids_scaled.index,
    columns=centroids_scaled.index,
)

sector_mix = pd.crosstab(analysis_df["cluster"], analysis_df["sector"], normalize="index")
sector_top3 = sector_mix.apply(lambda row: row.sort_values(ascending=False).head(3), axis=1)

def normalized_entropy(row: pd.Series) -> float:
    probs = row[row > 0].to_numpy()
    if len(probs) <= 1:
        return 0.0
    return float((-np.sum(probs * np.log(probs))) / np.log(len(row)))

def hhi_index(row: pd.Series) -> float:
    return float(np.sum(np.square(row.to_numpy())))

sector_concentration = pd.DataFrame({
    "entropy_norm": sector_mix.apply(normalized_entropy, axis=1),
    "hhi": sector_mix.apply(hhi_index, axis=1),
})

kruskal_rows: list[dict[str, float | str]] = []
for feature_name in clustering_columns:
    groups = [
        frame[feature_name].dropna().to_numpy()
        for _, frame in analysis_df.groupby("cluster")
    ]
    h_stat, p_val = stats.kruskal(*groups)
    n_obs = int(sum(len(g) for g in groups))
    k_groups = int(len(groups))
    eps_sq = (h_stat - k_groups + 1) / (n_obs - k_groups) if n_obs > k_groups else np.nan
    kruskal_rows.append({
        "feature": feature_name,
        "H": float(h_stat),
        "p_value": float(p_val),
        "epsilon_squared": float(eps_sq),
    })
kruskal_df = pd.DataFrame(kruskal_rows).sort_values(by="epsilon_squared", ascending=False)

ev_col = "Enterprise_Value_over_EBITDA"
valuation_df = analysis_df[["row_id", "cluster"]].merge(
    df_sectors_cook2[["row_id", ev_col]],
    on="row_id",
    how="left",
)
valuation_by_cluster = valuation_df.groupby("cluster")[ev_col].agg(["count", "mean", "median", "std"])

valuation_groups = [
    frame[ev_col].dropna().to_numpy()
    for _, frame in valuation_df.groupby("cluster")
    if frame[ev_col].notna().any()
]
valuation_kruskal = stats.kruskal(*valuation_groups) if len(valuation_groups) >= 2 else None

# Rule-based business labels for easier interpretation.
cluster_labels_business: list[dict[str, object]] = []
for cluster_id, row in cluster_profile_z.iterrows():
    tags: list[str] = []
    if row.get("Revenue_Growth", 0.0) >= 0.5:
        tags.append("growth-oriented")
    if row.get("Profit_Margin", 0.0) >= 0.5 and row.get("returnOnEquity", 0.0) >= 0.5:
        tags.append("high profitability")
    if row.get("Debt_to_Equity", 0.0) >= 0.5:
        tags.append("high leverage")
    if row.get("R_and_D_to_Revenue", 0.0) >= 0.5:
        tags.append("innovative (R&D)")
    if row.get("Capex_to_Revenue", 0.0) >= 0.5:
        tags.append("capital-intensive")
    if row.get("Profit_Margin", 0.0) <= -0.5:
        tags.append("margin pressure")
    if not tags:
        tags.append("mixed/stable profile")

    cluster_labels_business.append({
        "cluster": int(cluster_id),
        "label_business": "; ".join(tags),
    })
cluster_business_df = pd.DataFrame(cluster_labels_business)

print("=== QUALITY OF CLUSTER SEPARATION ===")
print({
    "silhouette_mean": round(overall_silhouette, 4),
    "calinski_harabasz": round(calinski_harabasz, 2),
    "davies_bouldin": round(davies_bouldin, 4),
})
print("\nSilhouette by cluster:")
print(silhouette_cluster_stats.round(4))

print("\n=== CLUSTER PROFILE (z-score vs global mean) ===")
print(cluster_profile_z.round(3))

print("\n=== KRUSKAL TEST ACROSS CLUSTERS (feature discrimination) ===")
print(kruskal_df.round(4))

print("\n=== SECTOR CONCENTRATION ===")
print(sector_concentration.round(4))
print("\nTop-3 sector shares per cluster:")
print(sector_top3.round(3))

print("\n=== VALUATION DIFFERENCES (EV/EBITDA) ===")
print(valuation_by_cluster.round(3))
if valuation_kruskal is not None:
    print(f"Kruskal EV/EBITDA: H={valuation_kruskal.statistic:.4f}, p={valuation_kruskal.pvalue:.6f}")
else:
    print("Kruskal EV/EBITDA skipped: insufficient non-null groups.")

print("\n=== BUSINESS LABELS (heuristic) ===")
print(cluster_business_df)

min_centroid_dist = float(np.min(centroid_distance.where(~np.eye(len(centroid_distance), dtype=bool)).stack()))
if overall_silhouette >= 0.25 and davies_bouldin <= 1.8 and min_centroid_dist >= 1.0:
    naturality_comment = "The cluster split looks relatively natural (good separation and meaningful centroid distance)."
elif overall_silhouette >= 0.18 and davies_bouldin <= 2.2:
    naturality_comment = "The split is moderately natural: visible groups exist, but some boundaries may be artificial."
else:
    naturality_comment = "The split may be partly artificial: low separation and/or strong cluster overlap."

print("\n=== NATURALNESS OF THE SPLIT (ECONOMIC-BUSINESS INTERPRETATION) ===")
print(naturality_comment)

plt.figure(figsize=(10, 4.5))
sns.heatmap(cluster_profile_z, annot=True, fmt=".2f", cmap="RdBu_r", center=0.0)
plt.title("Cluster profiles as z-scores vs global mean")
plt.xlabel("Financial ratio")
plt.ylabel("Cluster")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8.5, 4))
sns.barplot(
    x=sector_concentration.index.astype(str),
    y=sector_concentration["hhi"],
    color="#4C78A8",
)
plt.title("Sector concentration by cluster (HHI)")
plt.xlabel("Cluster")
plt.ylabel("HHI (higher = more concentrated)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8.5, 4))
sns.boxplot(data=analysis_df, x="cluster", y="silhouette_value", color="#72B7B2")
plt.axhline(0.0, color="black", linestyle="--", linewidth=1)
plt.title("Silhouette distribution by cluster")
plt.xlabel("Cluster")
plt.ylabel("Silhouette value")
plt.tight_layout()
plt.show()

plt.figure(figsize=(7.5, 5.5))
sns.heatmap(centroid_distance, annot=True, fmt=".2f", cmap="Blues")
plt.title("Pairwise centroid distances (scaled feature space)")
plt.xlabel("Cluster")
plt.ylabel("Cluster")
plt.tight_layout()
plt.show()
```

## Economic Differences Across Clusters: Synthesis of Results

### 1) Which differences are statistically strongest?

The strongest discriminating power comes from:

This means cluster membership is not merely a technical label; it carries real information about valuation levels.

### 3) Business interpretation cluster by cluster

Cluster 4 (innovation/growth, strong Technology + Healthcare tilt):
- Very high R&D, low leverage, and positive growth create a profile the market values as an option on future cash flows.
- This is the closest archetype to an "earlier stage or intense product/technology investment phase."
- In practice, investors accept higher short-term uncertainty in exchange for scaling potential.

Cluster 1 (capital-intensive and high-margin):
- Very high CAPEX with good margin and moderate ROE.
- Typical for asset-heavy models (e.g., real estate/infrastructure/energy), where growth is tied to investment cycles.
- Valuation is higher than in clusters 2 and 3, but lower than in innovation cluster 4.

Cluster 2 (high leverage + high ROE):
- High ROE coexists with the highest leverage.

Cluster 3 (large, diversified base):
- Largest sample share, weaker profitability and growth, low investment intensity and low R&D.

### 4) Naturalness of the split and practical implication

- Mean silhouette = 0.317 and Davies-Bouldin = 1.298 suggest the split is economically meaningful, though not perfectly "sharp."
- Cluster 3 is the most stable and homogeneous (high silhouette and no negative observations).
- Clusters 1 and 2 are more "borderline" (about 25-26% of observations with negative silhouette), which indicates partial profile overlap.

Practical conclusion: clusters should be treated as valuation/risk regimes, not rigid categories. The clearest market premium concentrates in the innovation cluster (4), while the neutral segment (3) remains the no-growth-premium reference point.
